```python
import pandas as pd
import pybedtools
```

```python
gene_meta = pd.read_csv(
    '/home/hanliu/ref/mouse/gencode/vm22/gencode.vM22.annotation.gene.flat.tsv.gz',
    index_col='gene_id',
    sep='\t')
```

```python
blacklist = pybedtools.BedTool('/home/hanliu/ref/blacklist/mm10-blacklist.v2.bed.gz')
gene_bed = pybedtools.BedTool().from_dataframe(gene_meta.reset_index()[['chrom', 'start', 'end', 'gene_id']])

gene_bed_white = gene_bed.intersect(blacklist, v=True)
white_genes = gene_bed_white.to_dataframe()
white_genes = pd.Index(white_genes['name'])
```

```python
good_genes = gene_meta.reindex(white_genes)
good_genes.index.name = 'gene_id'
good_genes.to_csv('gencode.vM22.annotation.gene.flat.filtered_white_genes.tsv.gz', sep='\t')
```
