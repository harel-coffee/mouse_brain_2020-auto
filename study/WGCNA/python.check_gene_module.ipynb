```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gene_module = pd.read_csv('gene_cluster.leiden.csv', index_col=0, squeeze=True)
gene_data = pd.read_csv('TotalPairwiseMarkerGene.csv', index_col=0)
gene_module=gene_module.sort_values()
gene_data = gene_data.loc[gene_module.index].copy()
```

```python
group_order = [
    'IT-L23+ORB', 'IT-L5+ORB', 'IT-L23+PFC', 'IT-L5+PFC', 'IT-L6+PFC',
    'IT-L23+ACA', 'IT-L5+ACA', 'IT-L6+ACA', 'IT-L23+AI', 'IT-L5+AI',
    'IT-L6+AI', 'IT-L23+MOs', 'IT-L4+MOs', 'IT-L5+MOs', 'IT-L6+MOs',
    'IT-L23+MOp', 'IT-L4+MOp', 'IT-L5+MOp', 'IT-L6+MOp', 'IT-L23+SSp',
    'IT-L4+SSp', 'IT-L5+SSp', 'IT-L6+SSp', 'IT-L23+SSs', 'IT-L4+SSs',
    'IT-L5+SSs', 'IT-L6+SSs'
]

gene_data = gene_data[group_order].copy()
```

```python
fig = plt.figure(figsize=(2, 4), dpi=200)
gs = fig.add_gridspec(1, 10)
ax1 = fig.add_subplot(gs[:, 0])
ax2 = fig.add_subplot(gs[:, 1:])

ax1.imshow(gene_module.values.reshape(gene_module.size, 1), 
          aspect='auto', cmap='rainbow')
ax2.imshow(gene_data.values, 
          aspect='auto', vmin=-1, vmax=1, cmap='coolwarm')
```
