## Gene meta

```python
gene_meta = pd.read_csv(
    '/home/hanliu/ref/mouse/gencode/vm22/gencode.vM22.annotation.gene.flat.tsv.gz',
    index_col='gene_id',
    sep='\t')
gene_meta.index.name = 'gene'
gene_name_to_id = {v: k for k, v in gene_meta['gene_name'].iteritems()}
gene_idbase_to_id = {i.split('.')[0]: i for i in gene_meta.index}
```

```python
def add_gene_data(gene):
    gene_id = gene_name_to_id[gene]
    if gene not in cell_tidy_data.columns:
        gene_data = mcds.sel(gene=gene_id, mc_type='CHN')['gene_da'].to_pandas()
        cell_tidy_data[gene] = gene_data
        
add_gene_data('Tle4')
add_gene_data('Bcl11b')
```

```python
gene = 'Bcl11b'

add_gene_data(gene=gene)

fig, ax = plt.subplots(figsize=(1.5, 1.5), dpi=1200)

continuous_scatter(ax=ax,
                   data=cell_tidy_data,
                   coord_base='l1-umap',
                   hue=gene,
                   hue_norm=(0, 2),
                   s=0.1,
                   max_points=None,
                   # text_anno='SubType', 
                   axis_format=None,
                   text_anno_kws=dict(fontsize=2), colorbar=False)

# add_ax_box(ax, linewidth=0.5)
fig.savefig(f'UMAP-{gene}.scatter_part.png')

fig, ax = plt.subplots(figsize=(0.5, 0.5), dpi=1200)

continuous_scatter(ax=ax,
                   data=cell_tidy_data,
                   coord_base='l1-umap',
                   hue=gene,
                   hue_norm=(0, 2),
                   s=0.3,
                   max_points=10000,
                   # text_anno='SubType', 
                   axis_format=None,
                   text_anno_kws=dict(fontsize=2))

ax.collections[0].remove()

add_ax_box(ax, linewidth=0.5)
fig.savefig(f'UMAP-{gene}.label_part.pdf')
```
