```python
import pandas as pd
import numpy as np
import joblib
```

```python
tf_genes = pd.read_csv('/home/hanliu/ref/TFGene/TFClass/TFClass.with_mouse_gene_id.csv', index_col=0)
unique_tf_genes = set([g for gs in tf_genes['EnsemblID'] if isinstance(gs, str) for g in gs.split(',')])
```

```python
gene_meta = pd.read_csv('/home/hanliu/ref/mouse/gencode/vm22/gencode.vM22.annotation.gene.flat.tsv.gz',
                        sep='\t', index_col='gene_id')
```

```python
region_markers = joblib.load('region_hypo_genes.IT-L23.obj')
```

```python
region_tf_markers = []
for region, gene_list in region_markers.items():
    genes = [g for g in gene_list if g.split('.')[0] in unique_tf_genes]
    df = gene_meta.loc[genes].copy()
    df['region'] = region
    region_tf_markers.append(df)
region_tf_markers = pd.concat(region_tf_markers)
```
