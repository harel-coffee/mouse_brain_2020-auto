```python
from papermill import execute_notebook
import joblib
import pandas as pd
import pathlib
```

## Layer Specific

```python
layer_specific_markers = pd.read_csv(
    '/home/hanliu/project/mouse_rostral_brain/study/ITSpatial/DMGAnalysis/LayerSpecificDMG.csv'
)
```

```python
# further filtering
layer_specific_markers = layer_specific_markers[
    (layer_specific_markers['AUROC'] > 0.9) & (layer_specific_markers['gene_mcg_from'] < 0.8) &
    (layer_specific_markers['gene_mcg_to'] > 1.2) &
    ((layer_specific_markers['gene_mcg_to'] - layer_specific_markers['gene_mcg_from']) > 0.75)].copy()
layer_specific_markers['gene_id'].unique().size
```

```python
for g in layer_specific_markers['gene_name'].unique():
    print(g)
    params = dict(gene=g, anno_col=0)
    if pathlib.Path(f'Genes/LayerSpecific/{g}.pdf').exists():
        continue

    input_path = 'layer_region_sunburst_gene.ipynb'
    output_path = 'Genes/LayerSpecific/layer_region_sunburst_gene.ipynb'
    execute_notebook(str(input_path),
                     str(output_path),
                     parameters=params,
                     engine_name=None,
                     prepare_only=False,
                     kernel_name=None,
                     progress_bar=False,
                     log_output=False,
                     start_timeout=60,
                     report_mode=False,
                     cwd=str('Genes/LayerSpecific/'))
```

## Region Specific

```python
region_specific_markers = pd.read_csv(
    '/home/hanliu/project/mouse_rostral_brain/study/ITSpatial/DMGAnalysis/RegionSpecificDMG.csv'
)
# further filtering
region_specific_markers = region_specific_markers[
    (region_specific_markers['AUROC'] > 0.9) & (region_specific_markers['gene_mcg_from'] < 1) &
    (region_specific_markers['gene_mcg_to'] > 1) &
    ((region_specific_markers['gene_mcg_to'] - region_specific_markers['gene_mcg_from']) > 0.5)].copy()
region_specific_markers['gene_id'].unique().size
```

```python
for g in region_specific_markers['gene_name'].unique():
    print(g)
    params = dict(gene=g, anno_col=1)
    if pathlib.Path(f'Genes/RegionSpecific/{g}.pdf').exists():
        continue

    input_path = 'layer_region_sunburst_gene.ipynb'
    output_path = 'Genes/RegionSpecific/layer_region_sunburst_gene.ipynb'
    execute_notebook(str(input_path),
                     str(output_path),
                     parameters=params,
                     engine_name=None,
                     prepare_only=False,
                     kernel_name=None,
                     progress_bar=False,
                     log_output=False,
                     start_timeout=60,
                     report_mode=False,
                     cwd=str('Genes/RegionSpecific/'))
```
