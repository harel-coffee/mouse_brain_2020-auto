## UMAP color by brain dissection region, label by major types

```python
major_region = 'CNU'
s = 0.5
```

```python
fig, ax = plt.subplots(figsize=(3, 3), dpi=1200)

categorical_scatter(ax=ax,
                    data=cell_tidy_data,
                    coord_base='l1-umap',
                    scatter_kws=dict(color='lightgray'),
                    s=s,
                    max_points=None)

_data = cell_tidy_data[cell_tidy_data['MajorRegion'] == major_region]
categorical_scatter(ax=ax,
                    data=_data.sample(_data.shape[0]),
                    coord_base='l1-umap',
                    hue='Region',
                    palette=region_palette,
                    s=s,
                    max_points=None,
                    #text_anno='MajorType', 
                    text_anno_kws=dict(fontsize=2))

add_ax_box(ax, linewidth=0.5)
pass
```
